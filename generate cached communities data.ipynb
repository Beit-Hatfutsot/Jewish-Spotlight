```python
import json

with open('bhjs-content/places/czech/cached-api.php') as f:
    cached_api = json.load(f)
```

```python
places = [item['Header']['En'] for item in cached_api['items'] if item['UnitTypeDesc'] == 'Place']
```

```python
import requests
import traceback

places_data = []

for place in places:
    try:
        data = requests.get(f'https://nominatim.openstreetmap.org/search?format=json&city={place}').json()[0]
        data = {'place': place, 'lat': data['lat'], 'lon': data['lon']}
        places_data.append(data)
        print(data)
    except Exception as e:
        print(f'{place}: {e}')
```

```python
places = {}
for place in places_data:
    places[place['place']] = place
```

```python
for item in cached_api['items']:
    if item['UnitTypeDesc'] == 'Place' and item['Header']['En'] in places:
        item['lat'] = place['lat']
        item['lon'] = place['lon']
```

```python
with open('bhjs-content/places/czech/cached-api.php', 'w') as f:
    json.dump(cached_api, f)
```
